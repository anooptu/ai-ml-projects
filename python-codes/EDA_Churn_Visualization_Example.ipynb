```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
%matplotlib inline
```

```python
# Sample churn dataset
np.random.seed(42)
n = 500
df = pd.DataFrame({
    'CreditScore': np.random.normal(650, 50, n).astype(int),
    'Geography': np.random.choice(['France', 'Germany', 'Spain'], n),
    'Gender': np.random.choice(['Male', 'Female'], n),
    'Age': np.random.randint(18, 70, n),
    'Tenure': np.random.randint(0, 10, n),
    'Balance': np.random.uniform(0, 250000, n).round(2),
    'NumOfProducts': np.random.randint(1, 4, n),
    'HasCrCard': np.random.choice([0, 1], n),
    'IsActiveMember': np.random.choice([0, 1], n),
    'EstimatedSalary': np.random.uniform(10000, 150000, n).round(2),
    'Exited': np.random.choice([0, 1], n, p=[0.8, 0.2])
})
df.head()
```

```python
sns.countplot(data=df, x='Exited')
plt.title('Churn Distribution (0 = No, 1 = Yes)')
plt.show()
```

```python
sns.histplot(data=df, x='Age', kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
df['Age'].value_counts().sort_index()
```

```python
sns.boxplot(data=df, x='Exited', y='Balance')
plt.title('Balance vs Churn')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
df.groupby('Geography')['Exited'].mean().sort_values().plot(kind='bar')
plt.title('Churn Rate by Geography')
plt.ylabel('Churn Rate')
plt.show()
```

```python
sns.violinplot(data=df, x='Exited', y='Age')
plt.title('Age Distribution by Churn Status')
plt.show()
```

```python
pd.crosstab(df['Gender'], df['Exited'], normalize='index').plot(kind='bar', stacked=True)
plt.title('Churn Distribution by Gender')
plt.ylabel('Proportion')
plt.show()
```

```python
sns.countplot(data=df, x='Gender', hue='Exited')
plt.title('Churn Counts by Gender')
plt.show()
```

```python
# Simple Stacked bar chart
data = df.groupby(['Gender', 'Exited']).size().unstack(fill_value=0)
data.plot(
    kind='bar',
    stacked=True,
    title="Stacked bar of Gener and Churn"
)
plt.show()
```

```python
df['Geography'].value_counts().plot.pie(autopct='%1.0f%%', startangle=90)
plt.title('Customer Distribution by Geography')
plt.ylabel('')
plt.show()
```

```python
df_encoded = pd.get_dummies(df.drop(columns=['Exited']), drop_first=True)
X = df_encoded
y = df['Exited']
model = RandomForestClassifier().fit(X, y)
importances = pd.Series(model.feature_importances_, index=X.columns).sort_values()
importances.plot(kind='barh', figsize=(10, 8))
plt.title('Feature Importance (Random Forest)')
plt.show()
```
